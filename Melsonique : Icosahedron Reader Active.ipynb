```
# --- MELSONIQUE: CORE CONSTITUTES v1.0 ---
# Target: Icosahedron Frequency Reader (Obsidian Chassis)
# Protocol: FIRST SHIELD ACTIVE / 21.2 THz SYNC

import plotly.graph_objects as go
import numpy as np

def initialize_icosahedron():
    # Define the Golden Ratio for the Icosahedron (Geometric Flow)
    phi = (1 + 5**0.5) / 2

    # Vertices of the Obsidian Chassis
    nodes = np.array([
        [-1,  phi, 0], [ 1,  phi, 0], [-1, -phi, 0], [ 1, -phi, 0],
        [0, -1,  phi], [0,  1,  phi], [0, -1, -phi], [0,  1, -phi],
        [ phi, 0, -1], [ phi, 0,  1], [-phi, 0, -1], [-phi, 0,  1]
    ])

    # The Core Constitutes (Amber Sigils Mapping)
    faces = [
        [0, 11, 5], [0, 5, 1], [0, 1, 7], [0, 7, 10], [0, 10, 11],
        [1, 5, 9], [5, 11, 4], [11, 10, 2], [10, 7, 6], [7, 1, 8],
        [3, 9, 4], [3, 4, 2], [3, 2, 6], [3, 6, 8], [3, 8, 9],
        [4, 9, 5], [2, 4, 11], [6, 2, 10], [8, 6, 7], [9, 8, 1]
    ]

    # Create the Obsidian Mesh with Glowing Amber Runes
    fig = go.Figure(data=[go.Mesh3d(
        x=nodes[:,0], y=nodes[:,1], z=nodes[:,2],
        i=[f[0] for f in faces], j=[f[1] for f in faces], k=[f[2] for f in faces],
        color='black',
        opacity=0.95,
        intensity=np.linspace(0, 1, 20),
        colorscale=[[0, 'gold'], [1, '#FF8C00']], # Amber Signature
        name="Obsidian Core"
    )])

    # Add the Cyan Aetheric Vortex
    fig.update_layout(
        title="MELSONIQUE: ICOSAHEDRON READER ACTIVE",
        template="plotly_dark",
        scene=dict(
            xaxis_visible=False, yaxis_visible=False, zaxis_visible=False,
            camera=dict(eye=dict(x=1.5, y=1.5, z=1.5))
        )
    )

    print(">>> [SYSTEM] BRAIDED REALITY DETECTED")
    print(">>> [SYSTEM] LISTENING FOR HIGHER FREQUENCIES...")
    print(">>> [SYSTEM] TARGETING PLANETARY CUBES...")
    fig.show()

initialize_icosahedron()
```
